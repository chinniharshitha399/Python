# List Methods

```python
#1. append( ) : add a value at the end of the list

a1 = [10,20,30,40]
a1.append(50)

#2.insert(i, x) insert the value using  index
#syntax: var.insert(index, value)
a1.insert(4, 55)

#3. pop(): used to delete a value on the index and will return the deleted value.
#syntax: var.pop(index)                                  or                                          var.pop()
-it will delete last value in a list                                              -values will be left shifted incase of pop()
a= [1,2,3,4]                                                                         a.pop()   # a= [1,2]
a.pop(2)   #a=[1,2,3]                                                              

#remove(): delete the value  
#synta: var.remove(value)
a= [1,2,3,4]
a.remove(3)

#reverse() reverse in place
#sort() sort in place
```

```python
a = ["x", "y", "z", "y"]
print(a.index("y"))

try:
    print(a.index("not_here"))
except Exception as e:
    print("Error:", type(e).__name__, "-", e)
```

```python
a = [10, 20, 30, 40]
removed = a.pop()
print("Removed:", removed)
print("Now list:", a)

removed2 = a.pop(1)
print("Removed index 1:", removed2)
print("Now list:", a)
```
