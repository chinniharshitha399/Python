# loops: 
Loops are used to repeat a block of code multiple times.

Types of Loops
There are two types of loops in python:
1. While
2. For

While Loop: Repeats the code as long as a condition is True.  

## Steps to run the While loop

1.Initializing the input Value
2.Creating while condition,to execute the loop until
3.Checks the condition
4.If satsifies
5.Print i vaue
6.If not terminates

Syntax:

while condition:
    # code block

```python
x = 3
while x > 0:
    print(x)
    x -= 1
```

```python
n = int(input("Enter a number: "))

print(f"\nLET'S LEARN THE TABLE OF {n} TODAY!")   #Headline

i = 1
while i<11:
    print(n,'*',i,'=',n*i)
    i+=1
```

# For Loop: 
-Used to iterate over sequences like lists, tuples, strings and ranges.

-performs the same operation to every item within the loop, it avoids managing index manually


```python
# Basic for loop over a list
numbers = [1, 2, 3, 4]
for n in numbers:
    print(n)
```

# Range function
range(6): 0,1,2,3,4,5    (starts from 0 by default)
range(3,8): 3,4,5,6,7
range(5,16,3): 5,8,11,14     (start, stop, step)

for i in range(3):
    print(i)

```python
# String:

for i in 'Python':
    print(i,end=" ")
```

```python
# Tuple:

for i in (3,5,7,10):
    print(i,end=" ")
```

```python
# List:

for i in [1,3,9]:
    print(i,end=" ")
```

```python
## Reversing a string (loop method): build a reversed string by adding each char to the front.
s = "picsart"
rev = ""

for i in range(len(s)):
    rev = s[i] + rev

print("Reversed:", rev)
```

```python
# Break and Continue Statement
1. # break —  Exit the loop completely
-Used to terminate a loop when a condition is met.
-It immediately exits the loop, even if the loop condition is still True.
```

```python
for i in range(5):
    if i == 3:
        break
    print(i, end=' ' )
```

# continue — 
 Skip the current iteration
-Skips the current loop iteration and jumps to the next one.
-The loop doesn’t stop, just avoids executing the rest of the code for that iteration.

```python
for i in range(5):
    if i == 3:
        continue
    print(i, end=' ' )
```
